```python
df = pd.read_csv('Student Mental health.csv')
df
```

```python
# Define a dictionary to map old column names to new column names
column_mapping = {'Choose your gender': 'Gender', 'What is your course?': 'Course', 'Your current year of Study' : 'Year', 'What is your CGPA?': 'CGPA', 'Do you have Depression?': 'Depression', 'Do you have Anxiety?': 'Anxiety', 'Do you have Panic attack?' : 'Panic_attack', 'Did you seek any specialist for a treatment?': 'Seeked_Specialist'}

df = df.rename(columns=column_mapping)
```

```python
df = df.dropna(subset = ['Age'])
```

```python
df['Age'].unique()
```

```python
# Loop through columns and convert text to lowercase
for col in df.columns:
    if df[col].dtype == 'object':
        df[col] = df[col].str.lower()
```

```python
# Columns to encode
columns_to_encode = ['Course', 'Year','Seeked_Specialist','Marital status', 'Depression', 'Anxiety', 'Panic_attack', 'Gender']

# Apply label encoding to selected columns
label_encoder = LabelEncoder()
for col in columns_to_encode:
    df[col] = label_encoder.fit_transform(df[col])

# Apply one-hot encoding to selected columns
#columns_to_one_hot_encode = ['Marital status', 'Depression', 'Anxiety', 'Panic_attack', 'Gender']
#df = pd.get_dummies(df, columns=columns_to_one_hot_encode)

df
```

```python
df['CGPA'].unique()
```

```python
def CGPA_function(x):
  if x == '3.00 - 3.49':
    return 3
  elif x == '3.50 - 4.00':
    return 4
  elif x == '2.50 - 2.99':
    return 2
  elif x =='2.00 - 2.49':
    return 1
  else:
    return 0
```

```python
df['CGPA'] = df['CGPA'].apply(CGPA_function)
df
```

```python
from sklearn.model_selection import train_test_split

# split the dataset into training and test sets in a 70:30 ratio
X_train, X_test, y_train, y_test = train_test_split(df.drop(["Timestamp", "Year", "CGPA","Marital status","Age","Course"], axis=1), df['Seeked_Specialist'], test_size=0.3, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
# Normalize or scale the data
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
from tensorflow import keras
from tensorflow.keras import layers
model = keras.Sequential()

# Add input layer
model.add(layers.Input(shape=(X_train.shape[1],)))

# Add one or more hidden layers
model.add(layers.Dense(64, activation='relu'))
model.add(layers.Dense(32, activation='relu'))

# Add the output layer with one neuron for regression
model.add(layers.Dense(1, activation='sigmoid'))

model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# Assuming you have one-hot encoded your target labels
# If your labels are not one-hot encoded, use 'sparse_categorical_crossentropy' as the loss function

model.fit(X_train, y_train, epochs=100, batch_size=32)

accuracy = model.evaluate(X_test, y_test, verbose=0)[1]
print(f'Accuracy: {accuracy * 100:.2f}%')
```

```python
from keras.metrics import Accuracy
# evaluate the model on the test data
loss, accuracy = model.evaluate(X_test, y_test)

# print the test loss and accuracy
print('Test loss:', loss)
print('Test accuracy:', accuracy)

# make predictions on the test data
y_pred = model.predict(X_test)

# round the predictions to the nearest integer
y_pred = np.round(y_pred)

# calculate the prediction accuracy

accuracy = Accuracy()
accuracy.update_state(y_pred, y_test)
prediction_accuracy = accuracy.result().numpy()

# print the prediction accuracy
print('Prediction accuracy:', prediction_accuracy)
```
